## **3. Fill Missing Values**

```python
plt.figure(figsize = (10, 5), facecolor = "white")
plot_num = 1
for i in ["avg_rating", "fast_charging"]:
    ax = plt.subplot(1,2, plot_num)
    sns.histplot(data = df, x = i, kde = True);
    plt.title(f"Data Distribution of {i}", fontsize = 12)
    plt.tight_layout()
    plot_num += 1

plt.show()
```

**Insight**

- The histogram above shows that 'avg_rating' column & 'fast_charging' column have a non-normal distribution (skewed). 
- Therefore, filling missing values with the median value.

### **Fill avg_rating**

```python
null_avg_rating = df[df["avg_rating"].isnull()]
null_avg_rating 
```

```python
fill_avg_rating = df.groupby("brand_name")[["avg_rating"]].median().reset_index()
fill_avg_rating 
```

```python
filled_avg_rating = pd.merge(null_avg_rating, fill_avg_rating, on = "brand_name", how = "left").set_index(null_avg_rating.index)
filled_avg_rating 
```

```python
filled_avg_rating.isna().sum()
```

```python
null_avg_rating["avg_rating"] = filled_avg_rating["avg_rating_y"]
```

```python
null_avg_rating.isna().sum()
```

```python
df[df["avg_rating"].isnull()] = null_avg_rating
```

```python
df.isna().sum()
```

### **Fill fast_charging**

```python
null_fast_charging = df[df["fast_charging"].isnull()]
null_fast_charging
```

```python
fill_fast_charging = df.groupby("brand_name")[["fast_charging"]].median().reset_index()
fill_fast_charging 
```

```python
filled_fast_charging = pd.merge(null_fast_charging, fill_fast_charging, on = "brand_name", how = "left").set_index(null_fast_charging.index)
filled_fast_charging 
```

```python
filled_fast_charging.isna().sum()
```

```python
null_fast_charging["fast_charging"] = filled_fast_charging["fast_charging_y"]
```

```python
null_fast_charging.isna().sum()
```

```python
df[df["fast_charging"].isnull()] = null_fast_charging
```

```python
df.isna().sum()
```

```python
df.dropna(inplace = True)
df.isna().sum()
```

## **4. Exploratory Data Analysis (EDA)**

```python
plt.figure(figsize = (10, 5))

plot = sns.histplot(data = df, x = "avg_rating", kde = True)
plot.set_title("Distribution of Average Ratings")
plot.set_xlabel("Average Ratings")
plot.set_ylabel("Frequency")

plt.show()
```

**Insight**
- Based on the histogram graph, it shows that the distribution of average ratings is most dominant in the range of 8 to 8.5.

```python
brand_rating = df.groupby("brand_name")[["avg_rating"]].mean().sort_values("avg_rating", ascending =  False).reset_index()
brand_rating
```

```python
# horizontal bar plot for average ratings on each brand

plt.figure(figsize = (15, 10))

plot = sns.barplot(data = brand_rating, x = "avg_rating", y = "brand_name", estimator = "mean", hue = "avg_rating")
plot.set_title("Average Ratings by Brand")
plot.set_xlabel("Average Ratings")
plot.set_ylabel("Brand Name")
plot.get_legend().remove()
for x, y in zip(brand_rating["avg_rating"], brand_rating["brand_name"]):
   plot.text(x = x + 0.1, y = y, s = "{:.1f}".format(x), color = "black", fontsize = 10, ha = "center")
   
plt.show()
```

```python
print(df.corr(numeric_only = True))
```

```python
# heatmap showing the correlation between each column

matrix = np.triu(df.corr(numeric_only = True))

plt.figure(figsize = (15, 10))

plot = sns.heatmap(data = df.corr(numeric_only = True),
                    annot = True,
                    cmap = "coolwarm",
                    center = 0,
                    mask = matrix,
                    linewidths = 1,
                    linecolor = "black")

plot.set_title("Correaltion between Each Column")

plt.show()
```

**Insight**
- From the heatmap above, it shows that 'avg_rating' has a strong positive correlation with 'ram_capacity', 'resolution_height', 'refresh_rate', 'processor_speed', and '5g_or_not', which means that those features have a strong impact on user ratings for smartphones.

```python
features = ["ram_capacity", "resolution_height", "refresh_rate", "processor_speed"]
target = "avg_rating"

plt.figure(figsize = (15, 10), facecolor = "white")
plot_num = 1

for feature in features:
    ax = plt.subplot(2,2, plot_num)
    sns.scatterplot(data = df, x =  feature, y = target, hue = target);
    plt.title(f"Correlation between {feature} and {target}", fontsize = 16)
    plt.tight_layout()
    plot_num += 1
```

```python
five_g = df.groupby("brand_name")[["5G_or_not"]].agg(lambda x: pd.Series.mode(x)[0]).sort_values("5G_or_not").reset_index()
five_g
```

```python
not_five_g = five_g[five_g["5G_or_not"] == 0]
not_five_g 
```

```python
not_five_g["brand_name"].values
```

**Insight**
- Brands such as itel, vivo, tecno, cola, doogee, tcl, nokia, huawei, infinix, oukitel, jio, lava, leeco, redmi, and letv do not offer 5G services.

```python
top10_brand_rating = brand_rating.head(10)
top10_brand_rating
```

```python
top10_brand_feature = df.groupby("brand_name")[["ram_capacity", "resolution_height", "refresh_rate", "processor_speed"]].mean().sort_values("ram_capacity", ascending =  False).reset_index().head(10)
top10_brand_feature
```

```python
top_brand = pd.merge(top10_brand_rating, top10_brand_feature, on = "brand_name", how = "inner")
top_brand
```

**Insight**
- By combining the 10 highest average ratings with the 10 highest features ('ram_capacity', 'resolution_height', 'refresh_rate', 'processor_speed'), similarities were found among 8 brands. Thus, it can be concluded that these features have an influence on user ratings for the type of smartphone used.

```python
bottom10_brand_rating = brand_rating.tail(10).reset_index().drop(columns = "index")
bottom10_brand_rating
```

```python
bottom10_brand_feature = df.groupby("brand_name")[["ram_capacity", "resolution_height", "refresh_rate", "processor_speed"]].mean().sort_values("ram_capacity", ascending =  False).reset_index().tail(10)
bottom10_brand_feature = bottom10_brand_feature.reset_index().drop(columns = "index")
bottom10_brand_feature
```

```python
bottom_brand = pd.merge(bottom10_brand_rating, bottom10_brand_feature, on = "brand_name", how = "inner")
bottom_brand 
```

**Insight**
- By combining the 10 lowest average ratings with the 10 lowest features ('ram_capacity', 'resolution_height', 'refresh_rate', 'processor_speed'), similarities were found among 8 brands. Thus, it can be concluded that these features have an influence on user ratings for the type of smartphone used.
